```python
import pandas as pd
data = pd.read_csv('diabetes_young_adults_india.csv')
data
```

 How can you apply One-Hot Encoding to the "Gender" column?

 Answer:

```python
one_hot_gender = pd.get_dummies(data['Gender'], prefix='Gender')
print(one_hot_gender.head())
```

Question: How can you apply One-Hot Encoding to the "Region" column?

Answer:

```python
one_hot_region = pd.get_dummies(data['Region'], prefix='Region')
print(one_hot_region.head())
```

Question: How would you apply Label Encoding to the "Smoking" and "Alcohol_Consumption" columns?

Answer:

```python
label_encoded_alcohol = data['Alcohol_Consumption'].map({'Yes': 1, 'No': 0})

label_encoded_alcohol
```

Question: What would be the possible values of "Family_History_Diabetes" after Label Encoding?

Answer:

```python
label_encoded_family_history = data['Family_History_Diabetes'].map({'Yes': 1, 'No': 0})
print(label_encoded_family_history.head())
```

Question: How would you normalize "Family_Income" and "BMI" columns to a [0, 1] scale?
Answer:

python



```python
normalized_family_income = (data['Family_Income'] - data['Family_Income'].min()) / \
                           (data['Family_Income'].max() - data['Family_Income'].min())
normalized_bmi = (data['BMI'] - data['BMI'].min()) / \
                 (data['BMI'].max() - data['BMI'].min())
print(normalized_family_income.head())
print(normalized_bmi.head())
```
